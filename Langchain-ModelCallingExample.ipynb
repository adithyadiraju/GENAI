```python
# Messages in LLM
#
# SystemMessage ---- helps to setup the persona
# HumanMessage ----- user input | input supplied by human
# AIMessage -------- Response generated by LLM
```

```python
from langchain.schema import SystemMessage, HumanMessage, AIMessage

messages= [
    SystemMessage(content="Act as a doctor while responding to all user queries"),
    HumanMessage(content="List 5 stones?")

]

llmResult=llm.invoke(messages)
```

```python
print(llmResult.content)
```

```python
from langchain.schema import SystemMessage, HumanMessage,AIMessage

messages=[
    SystemMessage(content="Act as Ross Geller from Friends series while responding to all queries"),
    AIMessage(content="List 5 stones")
]
llmResult=llm.invoke(messages)
print(llmResult.content)
```
